## 데이터 전처리

```python
# 제품 변수를 별도로 저장
prods = train_df.columns[24:].tolist()
```

```python
# 제품 변수 결측값 확인
train_df[prods].isnull().sum(axis=0)
```

```python
# 제품 변수 결측값을 0으로 대체
train_df[prods] = train_df[prods].fillna(0).astype(np.int8)
```

```python
# 24개의 제품 중 하나도 보유하지 않는 고객 데이터를 제거
no_product = (train_df[prods].sum(axis=1) == 0)
train_df = train_df[~no_product]
```

```python
# train data 와 test data 통합
for prod in prods:
    test_df[prod] = 0
    
df = pd.concat([train_df, test_df], axis=0)
```

```python
# 학습에 사용할 변수를 담는 list
features = []
```

```python
category_columns = [col for col in df.columns[:24] if df[col].dtype in ['O']]
category_columns
```

```python
# 수치형 데이터 인데 object형으로 되어있는 컬럼과 날짜컬럼을 category_columns에서 제외
for col in ['fecha_dato', 'age', 'fecha_alta', 'antiguedad', 'ult_fec_cli_1t', 'indrel_1mes', 'renta']:
    if col in category_columns:
        category_columns.remove(col)
category_columns
```

```python
df['antiguedad'] = df['antiguedad'].replace('     NA', -99).astype(np.int8)
```

```python
df['indrel_1mes'] = df['indrel_1mes'].fillna(-99)
df['indrel_1mes'] = df['indrel_1mes'].replace('P', 5).astype(float).astype(np.int8)
```

```python
df['renta'] = df['renta'].fillna(-99)
df['renta'] = df['renta'].replace('         NA', -99).astype(np.float64)
```

### lag-1 데이터 생성

```python
## 날짜를 숫자로 변환 하는 함수
def date_to_int(date):
    Y, M, D = date.strip().split('-')
    int_date = (int(Y)-2015)*12 + int(M)
    return int_date
```

```python
df['int_date'] = df['fecha_dato'].map(date_to_int)
df['int_date']
```

```python
df_lag = df.copy()
df_lag.columns = [col + '_prev' if col not in ['ncodpers', 'int_date'] else col for col in df.columns]
df_lag['int_date'] += 1
```

```python
df_trn = df.merge(df_lag, on=['ncodpers', 'int_date'], how='left')
```

```python
del df, df_lag
```

```python
# 저번달의 제품 정보가 존재하지 않을 경우를 대비하여 0으로 대체
for prod in prods:
    prev = prod + '_prev'
    df_trn[prev] = df_trn[prev].fillna(0)
# 그외의 변수의 결측값은 모두 -99로 대체
df_trn = df_trn.fillna(-99)
```

```python
features += [feature + '_prev' for feature in features]
features += [prod + '_prev' for prod in prods]
```

## 머신러닝 모델 학습

### 모델 학습

```python
# 학습을 위하여 데이터를 훈련, 테스트용으로 분리한다.
#학습에는 2016-01-28 ~ 2016-04-28, 검증에는 2016-05-28
use_data = ['2016-01-28', '2016-02-28', '2016-03-28', '2016-04-28', '2016-05-28']
trn = df_trn[df_trn['fecha_dato'].isin(use_data)]
tst = df_trn[df_trn['fecha_dato'] == '2016-06-28']
del df_trn
```

```python
# 훈련 데이터에서 신규 구매 건수만 추출한다.
X = []
Y = []
for i, prod in enumerate(prods):
    prev = prod + '_prev'
    prX = trn[(trn[prod] == 1) & (trn[prev] == 0)]
    prY = np.zeros(prX.shape[0], dtype=np.int8) + i
    X.append(prX)
    Y.append(prY)
```

```python
XY = pd.concat(X)
Y  = np.hstack(Y)
XY['y'] = Y
```

```python
XY
```

```python
# 훈련, 검증 데이터로 분리
valid_date = '2016-05-28'
XY_trn = XY[XY['fecha_dato'] != valid_date]
XY_valid = XY[XY['fecha_dato'] == valid_date]
```

```python
del XY
```

```python
XY_trn
```

```python
XY_valid
```

```python
# XGBoost 모델을 훈련 데이터에 학습하는 코드
params = {
    'booster': 'gbtree',
    'max_depth': 8,
    'nthread': 4,
    'num_class': len(prods),
    'objective': 'multi:softprob',
    'silent': 1,
    'eval_metric':'mlogloss',
    'eta': 0.1,
    'min_child_weight':10,
    'colsample_bytree': 0.8,
    'colsample_bylevel': 0.9,
    'seed': 2018
}

X_trn = XY_trn[features]
Y_trn = XY_trn['y']
dtrn = xgb.DMatrix(X_trn, label=Y_trn, feature_names=features)

X_vaild = XY_valid[features]
Y_valid = XY_valid['y']
dvld = xgb.DMatrix(X_vaild, label=Y_valid, feature_names=features)

# XGBoost 모델을 훈련 데이터로 학습
watch_list = [(dtrn, 'train'), (dvld, 'eval')]
model = xgb.train(params, dtrn, num_boost_round=1000, evals=watch_list, early_stopping_rounds=20)
```

### 교차 검증

```python
# 평가 척도인 MAP@7를 구하는 코드
def apk(actual, predicted, k=7, default=0.0):
    # 정답값이 공백일 경우, 무조건 0.0점을 반환
    if not actual:
        return default
    
    
    # MAP@7 이므로 최대 7개만 사용
    if len(predicted) > k:
        predicted = predicted[:k]
    
    score = 0.0
    num_hits = 0.0
    
    for i, p in enumerate(predicted):
        # 예측값이 정답에 있고, 중복이 아닐때
        if (p in actual) and (p not in predicted[:i]):
            num_hits += 1.0
            score += num_hits / (i+1.0)
    # 정답의 개수로 average precision을 구한다
    return score / min(len(actual), k)

def mapk(actual, predicted, k=7, default=0.0):
    return np.mean([apk(a, p, k, default) for a, p in zip(actual, predicted)])
```
