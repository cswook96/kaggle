## 머신러닝 파이프라인의 일반적인 순서
- 데이터 전처리 -> 피처 엔지니어링 -> 머신러닝 모델 학습 ->테스트 데이터 예측

## 데이터 전처리

```python
# 제품 변수를 별도로 저장
prods = train_df.columns[24:].tolist()
```

```python
# 제품 변수 결측값 확인
train_df[prods].isnull().sum(axis=0)
```

```python
# 제품 변수 결측값을 0으로 대체
train_df[prods] = train_df[prods].fillna(0).astype(np.int8)
```

```python
# 24개의 제품 중 하나도 보유하지 않는 고객 데이터를 제거
no_product = (train_df[prods].sum(axis=1) == 0)
train_df = train_df[~no_product]
```

```python
train_df
```

```python
# train data 와 test data 통합
for prod in prods:
    test_df[prod] = 0
    
df = pd.concat([train_df, test_df], axis=0)
```

```python
del train_df, test_df
```

```python
df.info()
```

```python
# 학습에 사용할 변수를 담는 list
features = []
```

```python
category_columns = [col for col in df.columns[:24] if df[col].dtype in ['O']]
category_columns
```

```python
# 수치형 데이터 인데 object형으로 되어있는 컬럼과 날짜컬럼을 category_columns에서 제외
for col in ['fecha_dato', 'age', 'fecha_alta', 'antiguedad', 'ult_fec_cli_1t', 'indrel_1mes', 'renta']:
    if col in category_columns:
        category_columns.remove(col)
category_columns
```

```python
# 범주형 변수를 factorize() 함수를 통해 label encoding
for col in category_columns:
    df[col], _ = df[col].factorize(na_sentinel=-99)
features += category_columns
```

```python
features
```

```python
# 수치형 변수의 특이값과 결측값을 -99로 대체하고, 수치형으로 변환
df['age'] = df['age'].replace(' NA', -99).astype(np.int8)
```

```python
df['antiguedad'] = df['antiguedad'].replace('     NA', -99).astype(np.int8)
```

```python
df['indrel_1mes'] = df['indrel_1mes'].fillna(-99)
df['indrel_1mes'] = df['indrel_1mes'].replace('P', 5).astype(float).astype(np.int8)
```

```python
df['renta'] = df['renta'].fillna(-99)
df['renta'] = df['renta'].replace('         NA', -99).astype(np.float64)
```

```python
df.info()
```

```python
# 학습에 사용할 수치형 변수를 features에 추가
features += ['age', 'antiguedad', 'ind_nuevo', 'indrel', 'indrel_1mes', 'ind_actividad_cliente']
features
```

## 피처 엔지니어링
- 머신러닝 모델 학습에 필요한 사용할 파생 변수를 생성한다.

```python
df.columns
```

```python
# 두 날짜 변수에서 연도와 월 정보를 추출
df['fecha_alta_year'] = df['fecha_alta'].map(lambda x:0.0 if x.__class__ is float else x.split('-')[0]).astype(np.int16)
df['fecha_alta_month'] = df['fecha_alta'].map(lambda x:0.0 if x.__class__ is float else x.split('-')[1]).astype(np.int16)
features += ['fecha_alta_year', 'fecha_alta_month']
```

```python
df['ult_fec_cli_1t_year'] = df['ult_fec_cli_1t'].map(lambda x:0.0 if x.__class__ is float else x.split('-')[0]).astype(np.int16)
df['ult_fec_cli_1t_month'] = df['ult_fec_cli_1t'].map(lambda x:0.0 if x.__class__ is float else x.split('-')[1]).astype(np.int16)
features += ['ult_fec_cli_1t_year', 'ult_fec_cli_1t_month']
```

```python
# 그외의 변수의 결측값은 모두 -99로 대체
df = df.fillna(-99)
```

### lag-1 데이터 생성

```python
## 날짜를 숫자로 변환 하는 함수
def date_to_int(date):
    Y, M, D = date.strip().split('-')
    int_date = (int(Y)-2015)*12 + int(M)
    return int_date
```

```python
df['int_date'] = df['fecha_dato'].map(date_to_int)
df['int_date']
```

```python
df_lag = df.copy()
df_lag.columns = [col + '_prev' if col not in ['ncodpers', 'int_date'] else col for col in df.columns]
df_lag['int_date'] += 1
```
