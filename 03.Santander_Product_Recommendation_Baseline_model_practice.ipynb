```python
# 훈련 데이터 로드
train_df = pd.read_csv('C:/Users/cswoo/Documents/kaggle_dataset/train_ver2.csv')
test_df = pd.read_csv('C:/Users/cswoo/Documents/kaggle_dataset/test_ver2.csv')
```

```python
# 24개의 고객 변수 
cus_va = train_df.columns[:24]
prod_va = train_df.columns[24:]
```

```python
# 제품 변수 null값 확인
train_df[prod_va].isnull().sum(axis=0)
```

```python
# 제품 변수 null값은 0으로 대치(null값은 제품을 보유하지 않은것으로 간주)
train_df[prod_va] = train_df[prod_va].fillna(0).astype(np.int8)
```

```python
# 두 데이터 프레임을 결합(전처리를 한번에 해주기 위해서)
df = pd.concat([train_df, test_df], axis=0)

# 메모리 효율을 위해 train_df, test_df 삭제
del train_df, test_df
```

```python
df
```

```python
# 고객변수의 dtype과 고유값 확인
for col in cus_va:
    print(col)
    print(df[col].dtype)
    print(df[col].unique())
    print('-'*50)
```

```python
non_features = ['fecha_dato', 'ncodpers', 'fecha_alta', 'ult_fec_cli_1t']
features = []
for cus in cus_va:
    if cus not in non_features:
        features.append(cus)
# 학습에 사용할 피쳐
features
```

```python
# 범주형 데이터 컬럼
category_cols = ['ind_empleado', 'pais_residencia', 'sexo', 'tiprel_1mes', 'indresi', 'indext', 'conyuemp', 'canal_entrada', 'indfall', 'nomprov', 'segmento']
```

```python
# 수치형인데 범주형으로 되어있는 컬럼
ca2num_cols = ['age', 'antiguedad', 'indrel_1mes', 'renta']
```

```python
for col in category_cols:
    df[col] = df[col].factorize(na_sentinel=-99)[0]
    df[col] = df[col].replace(-99, np.nan)
```

```python
df['age'] = df['age'].replace(' NA', np.nan).astype(np.float16)
```

```python
df['antiguedad'] = df['antiguedad'].replace({'     NA':np.nan, '-999999':np.nan, -999999:np.nan}).astype(np.float16)
```

```python
df['indrel_1mes'] = df['indrel_1mes'].replace('P', 5).astype(np.float16)
```

```python
df['renta'] = df['renta'].replace('         NA', np.nan).astype(np.float64)
```

```python
# 전처리를 완료 했으므로 다시 train_df와 test_df로 나누기
train_df = df[df['fecha_dato'] != '2016-06-28']
test_df = df[df['fecha_dato'] == '2016-06-28'].drop(prod_va, axis=1)
```

```python
# 메모리 효율을 위해 df 삭제
del df
```

```python
def date2int(date):
    Y, M, D = date.split('-')
    int_date = (int(Y)-2015)*12 + int(M)
    return int_date
```

```python
train_df['int_date'] = train_df['fecha_dato'].map(date2int)
```

```python
# lag1 데이터 만들기
train_df_lag1 = train_df.copy()
train_df_lag1['int_date'] += 1
```

```python
# train_df와 train_df_lag1 merge
df_train = train_df.merge(train_df_lag1, how='left', on=['ncodpers', 'int_date'])

# 메모리를 위해 train_df, train_df_lag1 데이터 삭제
del train_df, train_df_lag1
```
